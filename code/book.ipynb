```python
table = driver.find_elements(By.CSS_SELECTOR, '#content > div.post > div > div:nth-child(2) > table > tbody')
len(table)
```

```python
print(table[0].text)
```

```python
rows = table[0].find_elements(By.TAG_NAME, 'tr')
print(rows[1].text)
```

```python
data_fr = []

for i in range(28):
    
    table = driver.find_elements(By.CSS_SELECTOR, '#content > div.post > div > div:nth-child(2) > table > tbody')

    rows = table[0].find_elements(By.TAG_NAME, 'tr')

    for f in rows:
        
        elems = f.find_elements(By.TAG_NAME, 'td')
        
        temp = []
        
        for e in elems:
            temp.append(e.text)
            
        data_fr.append(temp)

    sig = driver.find_element(By.CSS_SELECTOR, '#content > div:nth-child(2) > span > a')

    time.sleep(3)

    sig.click()    
    
print(data_fr)
```
