```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
data_circuits = pd.read_csv('../../data-projects/data_f1/circuits.csv', encoding = 'utf-8', encoding_errors = 'ignore')
data_races = pd.read_csv('../../data-projects/data_f1/races.csv', encoding = 'utf-8', encoding_errors = 'ignore')
```

```python
data_circuits.shape
```

```python
data_circuits.head()
```

```python
data_circuits = data_circuits.drop(['circuitRef', 'lat', 'lng', 'alt', 'url'], axis = 1)
data_circuits.head()
```

```python
data_circuits = data_circuits.rename(columns={'circuitId': 'circuit_id'})
data_circuits.head()
```

```python
data_races.shape
```

```python
data_races.head()
```

```python
data_races = data_races.drop(['fp1_date', 'fp1_time', 'fp2_date', 'fp2_time', 'fp3_date', 'fp3_time', 'quali_date', 'quali_time', 'sprint_date', 'sprint_time', 'url'], axis = 1)
```

```python
data_races.head()
```

```python
data_races = data_races.rename(columns={'raceId': 'race_id', 'round': 'race_num', 'circuitId': 'circuit_id'})
data_races.head()
```

```python
data_races.year.unique()
```

```python
data_races = data_races.sort_values(by = 'date', ascending = False)
data_races.head()
```

```python
data_races = data_races.reset_index(drop = True)
```

```python
data_races.head()
```

```python
# unir los dataframes
data_merged = pd.merge(data_races, data_circuits, on = 'circuit_id')
data_merged.head()
```

```python
data_merged = data_merged.sort_values(by = 'date', ascending = False)
data_merged.head()
```

```python
data_merged = data_merged.reset_index(drop = True)
data_merged.head()
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options

import time
import warnings
warnings.filterwarnings('ignore')

# opciones del driver
opciones = Options()

# quita la bandera de ser robot
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)

# guardar las cookies
opciones.add_argument('user-data-dir=cookies')    # mantiene las coockies
```

```python
PATH = ChromeDriverManager().install()

driver = webdriver.Chrome(PATH)

driver.get('https://f1.fandom.com/wiki/Safety_Car')
```
