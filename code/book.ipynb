```python
PATH = ChromeDriverManager().install()

driver = webdriver.Chrome(PATH)

driver.get('https://f1.fandom.com/wiki/Safety_Car')
```

```python
aceptar = driver.find_element(By.CSS_SELECTOR, 'body > div:nth-child(18) > div > div > div._1r08nyekFdI7_2d8r3AIBf > div.NN0_TB_DIsNmMHgJWgT7U.XHcr6qf5Sub2F2zBJ53S_')

time.sleep(3)

aceptar.click()
```

```python
tabla = driver.find_elements(By.TAG_NAME, 'table')
len(tabla)
```

```python
print(tabla[1].text)
```

```python
filas = tabla[1].find_elements(By.TAG_NAME, 'tr')
print(filas[0].text)
```

```python
data = []


for f in filas:
    
    elems = f.find_elements(By.TAG_NAME, 'td')
    
    tmp = []
    
    for e in elems:
        tmp.append(e.text)
        
    data.append(tmp)
    
    
print(data)
```

```python
cab = tabla[1].find_element(By.TAG_NAME, 'tr')

texto = cab
tempo = texto.text.split()
print(tempo)
```

```python
df = pd.DataFrame(data)

df.head(15)
```

```python
new_col = ['Grand Prix', 'Cause', 'Deployed (lap)', 'Called in (lap)', 'Number of full laps']
df.columns = new_col

df.head(15)
```

```python
df = df.drop(0).reset_index(drop=True)
```

```python
for i in range(len(df['Grand Prix'])):
    if df['Grand Prix'][i] is not None and 'Grand Prix' not in df['Grand Prix'][i]:
        df['Number of full laps'][i] = df['Called in (lap)'][i]
        df['Called in (lap)'][i] = df['Deployed (lap)'][i]
        df['Deployed (lap)'][i] = df['Cause'][i]
        df['Cause'][i] = df['Grand Prix'][i]
        df['Grand Prix'][i] = df['Grand Prix'][i-1]
```

```python
df.tail()
```

```python
# Seleccionar la fila con índice 296
fila_297 = df.loc[297]

# Modificar la información de la fila a tu gusto
fila_297['Grand Prix'] = '2022 Emilia Romagna Sprint'
fila_297['Cause'] = 'Accident (Zhou)'
fila_297['Deployed (lap)'] = 1
fila_297['Called in (lap)'] = 4
fila_297['Number of full laps'] = 2

# Asignar la fila modificada de vuelta al DataFrame
df.loc[297] = fila_297
```

```python
df['year'] = df['Grand Prix'].str.extract('(\d{4})')
df.head()
```

```python
df['Grand Prix'] = df['Grand Prix'].str.replace(r'^\d{4}\s', '', regex=True)
df.head()
```
