```python
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

import time

import warnings
warnings.filterwarnings('ignore')
```

```python
#driver configuration
#opciones.add_argument('--start-maximized')         # comienza maximizado
#opciones.add_extension('driver_folder/adblock.crx')       # adblocker
#opciones.add_argument('user-data-dir=selenium')    # mantiene las cookies
#opciones.add_argument('--incognito')

opciones = Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.headless = False
```

```python
PATH = Service(ChromeDriverManager(version="114.0.5735.90").install())

driver = webdriver.Chrome(service=PATH, options=opciones)

driver.get('https://f1.fandom.com/wiki/Safety_Car')
```

```python
data = []


for f in filas:
    
    elems = f.find_elements(By.TAG_NAME, 'td')
    
    tmp = []
    
    for e in elems:
        tmp.append(e.text)
        
    data.append(tmp)
    
    
data[:10]
```

```python
driver.quit()
```

```python
new_col = ['grand_prix', 'cause', 'deployed_lap', 'called_in_lap', 'number_of_full_laps']
safety_car.columns = new_col

safety_car.head()
```

```python
safety_car = safety_car.drop(0).reset_index(drop=True)
```

```python
for i in range(len(safety_car['grand_prix'])):
    if safety_car['grand_prix'][i] is not None and 'Grand Prix' not in safety_car['grand_prix'][i]:
        safety_car['number_of_full_laps'][i] = safety_car['called_in_lap'][i]
        safety_car['called_in_lap'][i] = safety_car['deployed_lap'][i]
        safety_car['deployed_lap'][i] = safety_car['cause'][i]
        safety_car['cause'][i] = safety_car['grand_prix'][i]
        safety_car['grand_prix'][i] = safety_car['grand_prix'][i-1]
```

```python
safety_car.tail()
```

```python
fila_297 = safety_car.loc[297]

fila_297['grand_prix'] = '2022 Emilia Romagna Sprint'
fila_297['cause'] = 'Accident (Zhou)'
fila_297['deployed_lap'] = 1
fila_297['called_in_lap'] = 4
fila_297['number_of_full_laps'] = 2

safety_car.loc[297] = fila_297
```

```python
safety_car['year'] = safety_car['grand_prix'].str.extract('(\d{4})')
safety_car.head()
```

```python
safety_car['grand_prix'] = safety_car['grand_prix'].str.replace(r'^\d{4}\s', '', regex=True)
safety_car.head()
```

```python
safety_car['grand_prix'].unique()
```

```python
safety_car = safety_car.sort_values(by = 'year', ascending = False)
safety_car = safety_car.reset_index(drop = True)

safety_car.head()
```

```python
grand_prix_mapping = races.set_index('grand_prix')['circuitId'].to_dict()
safety_car['circuitId'] = safety_car['grand_prix'].map(grand_prix_mapping)

safety_car['circuitId'] = safety_car['circuitId'].astype(float).astype('Int64')

safety_car['circuitId'] = safety_car['circuitId'].astype(object)
safety_car['circuitId'] = safety_car['circuitId'].fillna('\\N')

safety_car.head()
```

```python
PATH = Service(ChromeDriverManager(version="114.0.5735.90").install())

driver = webdriver.Chrome(service=PATH, options=opciones)

driver.get('http://ergast.com/api/f1/drivers?=123')
```

```python
data_fr = []

for i in range(28):
    
    table = driver.find_elements(By.CSS_SELECTOR, '#content > div.post > div > div:nth-child(2) > table > tbody')

    rows = table[0].find_elements(By.TAG_NAME, 'tr')

    for f in rows:
        
        elems = f.find_elements(By.TAG_NAME, 'td')
        
        temp = []
        
        for e in elems:
            temp.append(e.text)
            
        data_fr.append(temp)

    sig = driver.find_element(By.CSS_SELECTOR, '#content > div:nth-child(2) > span > a')

    time.sleep(3)

    sig.click()    
    
data_fr[:10]
```
